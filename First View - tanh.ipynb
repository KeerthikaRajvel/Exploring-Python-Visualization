```python
import matplotlib.pyplot as plt
#to see plots inside the Jupyter notebook
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(-100,100)
y = np.tanh(x)
```

```python
#Functional method of creating a plot
plt.plot(x,y,'r-')
plt.xlabel('Input data')
plt.ylabel('tanh value')
plt.title('tanh plot')
```

```python
#To create multiple plots
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b-')
#tight_layout is used to ensure the axis of one graph doesn't overlap with the other - spacing/padding between graphs
plt.tight_layout()
```

```python
#Object-oriented method of creating a plot
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])
ax.plot(x,y)
ax.set_xlabel('Input data')
ax.set_ylabel('tanh value')
ax.set_title('tanh plot')
```

```python
#To create multiple plots
f = plt.figure()
ax1 = f.add_axes([0.1,0.1,1,1])
ax2 = f.add_axes([0.17,0.4,0.4,0.3])
ax1.plot(x,y,'r-')
ax1.set_title('Large')
ax2.plot(y,x)
ax2.set_title('Small')
```

```python
#Creating subplots
"""Can Specify number of columns and rows with subplots and get plots next to each other instead of one within the other.
Note: Axes is just an array of all the axes. We can even loop through it and display plots"""
fig, axes = plt.subplots(nrows=1, ncols=2)
'''for axis in axes:
    axis.plot(x,y)'''
#Can we index it individually
axes[0].plot(x,y,'r-')
axes[0].set_title('First plot')
axes[1].plot(x,y,'g-')
axes[1].set_title('Second plot')
for axis in axes:
    axis.set_xlabel('Input data')
    axis.set_ylabel('tanh value')
plt.tight_layout()
```

Figure Size, Aspect Ratio and DPI

```python
"""figsize attribute inside plt.figure takes width and height of the plot in inches. 
dpi is the dots per inch or pixels per inch. Both of these are optional attributes"""
fig = plt.figure(figsize=(4,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(4,2),dpi=100)
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
ax[0].plot(x,y)
ax[1].plot(y,x)
plt.tight_layout()
```

```python
#Plotting 2 or more curves in the same plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x vs. y')
ax.plot(y,x,label='y vs. x')
#loc is used to specify location of the legend. 0 - best, 1 - upper right, 2 - upper left, etc. Can customize it specifying position in ()
ax.legend(loc=0)
#ax.legend(loc=(0,0.1))
```

Plot Apperance

```python
#Color parameter. Can give common color names, RGB and HEX codes
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#40E0D0')
```

```python
#Linewidth parameter - default is 1; alpha parameter - used to specify transparency of line; lw - linewidth shorthand
#Linestyle - ls is the shorthand; marker
x = np.linspace(0,5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='green',lw=5,alpha=0.5,ls='--',marker='o',markersize=5,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='red')
ax.set_xlim([0,1])
ax.set_ylim([0,10])
```
