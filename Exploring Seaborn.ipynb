Distribution Plot

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
'''Dist plot shows the distribution of a univariate(one variable) set of observations'''
sns.distplot(tips['total_bill']) 
'''Here we get a histogram and a KDE which is a Kernal Density Estimation'''
```

```python
sns.distplot(tips['total_bill'],kde=False) #Without KDE
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30) #Without KDE and customizing number of bins
```

Joint Plot

```python
'''It is used to combine 2 different variables ie, combine 2 distribution plots'''
sns.jointplot(x='total_bill',y='tip',data=tips) #Kind parameter is used to customize the inside of the plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind = 'hex' ) #Kind parameter is used to customize the inside of the plot
```
